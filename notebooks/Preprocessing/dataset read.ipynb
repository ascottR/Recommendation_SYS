```python
def remove_musical_instruments(cat_string):
    try:
        # Try to convert the string to a list
        cat_list = ast.literal_eval(cat_string)
        # If it's a list, remove 'Musical Instruments'
        if isinstance(cat_list, list):
            cleaned_list = [cat for cat in cat_list if cat != 'Musical Instruments']
            return cleaned_list
    except (ValueError, SyntaxError):
        # If it's not a valid list, just return the original string
        return cat_string

# Apply the function to the 'categories' column
df.loc[:,'categories'] = df['categories'].apply(remove_musical_instruments)

df
```

```python
# Get the count of each category
categories_count = df['categories'].value_counts()

# Get categories with counts less than 2
cat46 = categories_count[categories_count < 2]

# Create a DataFrame from cat46 to join with the parent_asin
cat46_df = pd.DataFrame(cat46).reset_index()
cat46_df.columns = ['categories', 'count']

# Merge with the original DataFrame to get parent_asin values
result = pd.merge(cat46_df, df[['parent_asin', 'categories']], on='categories', how='left')

# Drop duplicates to only show unique pairs
result = result.drop_duplicates()

print(result)
```

```python
# Create 'sec_cat' only for rows where 'categories_tuple' has at least 2 elements
df.loc[:, 'sec_cat'] = df['categories_tuple'].apply(lambda x: x[1] if len(x) > 1 else None)

# Drop rows where 'sec_cat' is None
df = df.dropna(subset=['sec_cat'])

# Count the occurrences of each category in the 'first_cat' column
categories_count = df['sec_cat'].value_counts()

# Filter categories with more than one occurrence
cat46 = categories_count[categories_count > 1]

# Print the result
print(cat46)
print("second category count is:",cat46.count())
```

```python
# Drop the 'first_cat' and 'morethan1cat' columns
df = df.drop(['first_cat', 'morethan1cat'], axis=1)

# Check the DataFrame to ensure the columns were dropped
print(df.columns)
```

```python
# Count the occurrences of each category
categories_count = df['sec_cat'].value_counts()

# Filter out categories with less than 2 items
filtered_df = df[df['sec_cat'].isin(categories_count[categories_count > 1].index)]
```

```python
from sklearn.model_selection import train_test_split

# Perform stratified sampling with 50% of the filtered data
df_sample, _ = train_test_split(filtered_df, test_size=0.5, stratify=filtered_df['sec_cat'], random_state=42)

# Check the number of items sampled from each category
print(df_sample['sec_cat'].value_counts())
```
